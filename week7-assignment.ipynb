```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("movie_ratings.csv")
```

```python
movies.head()
```

```python
movies_long = pd.melt(movies, id_vars = ["Movies"], value_vars = ["P1", "P2", "P3", "P4", "P5"])
```

```python
movies_long.head()
```

```python
movies_long.columns = ["Movies", "Person", "Rating"]
```

```python
movies_long.head()
```

```python
movies_long.groupby(["Person"])[["Rating"]].mean()
```

```python
movies_long.groupby(["Movies"])[["Rating"]].mean()
```

```python
movies_long["Normalized Rating"] = movies_long.groupby('Person')["Rating"].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
movies_long_normalized = movies_long[["Movies", "Person", "Normalized Rating"]]
```

```python
movies_long_normalized
```

```python
movies_long_normalized.groupby(["Person"])[["Normalized Rating"]].mean().sort_values("Normalized Rating", ascending=False)
```

```python
movies_long.groupby(["Person"])[["Rating"]].mean().sort_values("Rating", ascending=False)
```

```python
movies_long_normalized.groupby(["Movies"])[["Normalized Rating"]].mean().sort_values("Normalized Rating", ascending=False)
```

```python
movies_long.groupby(["Movies"])[["Rating"]].mean().sort_values("Rating", ascending=False)
```

#### we have noticed that after normalization of the ratings for person, p5 has moved from being top rater(avg 3.5) to lowest rater(avg 0.25 on normalized scale). Normalization is helpful in this case because different person can give rating of the movies differently. One person can be very harsh while rating the movies while another can easily give higher rating to movies. So normalizing the rating brings the rating on a similar scale and therefore removes any personal bias. One disadvantage is that since we are transforming the number on different scales, it might not make intuitive sense to us as we are more used to 0-5 scales rating instead of 0-1. 

```python
movies_long["Standardized Rating"] = movies_long.groupby('Person')["Rating"].transform(lambda x: (x - x.mean()) / x.std())
```

```python
movies_long_standardized = movies_long[["Movies", "Person", "Standardized Rating"]]
```

```python
movies_long_standardized
```

```python
movies_long_standardized.groupby(["Movies"])[["Standardized Rating"]].mean()
```

```python
movies_long_standardized.groupby(["Person"])[["Standardized Rating"]].mean()
```
